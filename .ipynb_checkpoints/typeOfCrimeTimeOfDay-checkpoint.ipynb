```python
import numpy as np
import pandas as pd
import csv
import datetime as dtt
import matplotlib.pyplot as plt
from datetime import datetime
```

```python
file = pd.read_csv("crime_data_log.csv")

df = pd.DataFrame()
df['time'] = file['Time Reported']
df['type'] = file['Type']

df
```

```python
df['time'] = pd.to_datetime(df['time'], format='%I:%M %p', errors='coerce').dt.time
```

```python
hours = {}
typeOfCrimePerHour = {}

for i in df['time']:
    for hour in range(23):
        if i > dtt.time(hour,0,0) and i < dtt.time(hour + 1,0,0):
            if hour in hours:
                hours[hour] += 1
                typeOfCrimePerHour[hour] = df.loc[df.time == hour]
            else:
                hours[hour] = 1
```

```python
plt.figure(figsize=(20,10))
plt.xlabel('Hour')
plt.ylabel('Number of Crimes')
plt.title('Number of Crimes Over Hour in a Day\nHavard Yard Area')
plt.bar(list(hours.keys()), hours.values(), color='g')
plt.show()
```
